```python
import pandas as pd
df = pd.read_csv('/Users/navyasinghal/Desktop/Data Science Programming/Final Project/train.csv')
df = df.drop(df.iloc[:,[0]], axis=1)
df.set_index('id', inplace = True)
df.head()
#df.dtypes
#df.isnull().sum()
#df.shape
```

```python
categorical_indexes = list(range(6,20))
df.iloc[:,categorical_indexes] = df.iloc[:,categorical_indexes].astype('category')
df.dtypes
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.head()
```

```python
df['Arrival_Delay_in_Minutes'] = df['Arrival_Delay_in_Minutes'].fillna(df['Arrival_Delay_in_Minutes'].mean())
df.head()
```

```python
import matplotlib.pyplot as plt

# Filter the data based on customer type
loyal_customers = df[df['Customer_Type'] == 'Loyal Customer']
disloyal_customers = df[df['Customer_Type'] == 'disloyal Customer']

# Calculate satisfaction levels
loyal_satisfaction = loyal_customers['satisfaction'].value_counts(normalize=True) * 100
disloyal_satisfaction = disloyal_customers['satisfaction'].value_counts(normalize=True) * 100

# Create a DataFrame for plotting
satisfaction_df = pd.DataFrame({
    'Loyal Customer': loyal_satisfaction,
    'Disloyal Customer': disloyal_satisfaction
}).T

# Plot the bar chart
satisfaction_df.plot(kind='bar', figsize=(10, 6))
plt.title('Satisfaction Levels of Loyal vs Disloyal Customers')
plt.xlabel('Customer Type')
plt.ylabel('Percentage')
plt.legend(title='Satisfaction Level')
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the counts of satisfaction levels
loyal_satisfaction_count = loyal_customers['satisfaction'].value_counts()
disloyal_satisfaction_count = disloyal_customers['satisfaction'].value_counts()

# Create a DataFrame for plotting
satisfaction_count_df = pd.DataFrame({
    'Loyal Customer': loyal_satisfaction_count,
    'Disloyal Customer': disloyal_satisfaction_count
}).T

# Plot the bar chart with counts
ax = satisfaction_count_df.plot(kind='bar', figsize=(10, 6), stacked=True)
plt.title('Satisfaction Levels of Loyal vs Disloyal Customers')
plt.xlabel('Customer Type')
plt.ylabel('Count')
plt.legend(title='Satisfaction Level')
plt.xticks(rotation=0)

# Display the count of users on the bar plot
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() * 1.005, p.get_height() * 1.005))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define flight distance bins according to industry norms
bins = [0, 2000, float('inf')]
labels = ['Short-haul', 'Long-haul']

# Create a new column for binned flight distance
df['Distance Category'] = pd.cut(df['Flight_Distance'], bins=bins, labels=labels)

# Group by distance category and satisfaction, then count the occurrences
satisfaction_counts = df.groupby(['Distance Category', 'satisfaction'], observed=True).size().unstack()

# Normalize the counts to percentages
satisfaction_percentages = satisfaction_counts.div(satisfaction_counts.sum(axis=1), axis=0) * 100

# Plot the normalized bar plot
satisfaction_percentages.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Customer Satisfaction by Flight Distance Category (Normalized)')
plt.xlabel('Flight Distance Category')
plt.ylabel('Percentage of Customers')
plt.legend(['Dissatisfied', 'Satisfied'])
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots
f, ax = plt.subplots(1, 2, figsize=(15, 5))

# Boxplot with hue to avoid deprecation warning
sns.boxplot(x="Customer_Type", y="Age", hue="Customer_Type", data=df, ax=ax[0], legend=False)

# Histogram plot
sns.histplot(df, x="Age", hue="Customer_Type", multiple="stack", edgecolor=".3", linewidth=.5, ax=ax[1])

plt.show()
```

From this box diagram, we can conclude that most of the airline's regular customers are between the ages of 30 and 50 (their average age is slightly over 40). The age range of non-regular customers is slightly smaller (from 25 to 40 years old, on average - a little less than 30).
